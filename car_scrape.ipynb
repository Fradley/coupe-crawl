```python
from selenium.webdriver import Firefox
import pandas as pd
import time
import numpy as np
import time
```

```python
driver = Firefox()
page = 1
cars = list()
```

```python
t0 = time.time()
while True:
    t = time.time()
    driver.get(f'https://www.cars.com/for-sale/searchresults.action/?bsId=20203&dealerType=localOnly&page={page}&perPage=100&searchSource=GN_BREADCRUMB&sort=relevance&zc=30318')
    time.sleep(4)
    counter = 100
    car_ad_elements = driver.find_elements_by_class_name('listing-row__details')
    car_text = [(i.text).split('\n')[:8] for i in car_ad_elements]
    del_list = list()
    for j in range(len(car_text)):
        i = car_text[j]
        try:
            if i[3].isupper():
                del i[3]
            if i[-1] == 'Save':
                del i[-1]
            data = i[1].split(' ')
            i += [data[0]]
            i += [data[1]]
            del i[1]

            i[2] = i[2].split(': ')[1]
            i[3] = i[3].split(': ')[1]
            i[4] = i[4].split(': ')[1]
            i[5] = i[5].split(': ')[1]
            i[6] = i[6][1:].replace(',', '')
            i[7] = i[7].replace(',', '')
        except:
            del_list.append(j)
            counter -= 1

    for i in reversed(del_list):
        del car_text[i]

    cars += car_text
    elapsed = time.time() - t
    overall = time.time() - t0
    print(f'\rCurrently scraping page: {page}\n\tCurrently have {len(cars)} cars\n\tThis page took: {elapsed:.2f} s\n\tRunning for: {overall / 60 : .2f} minutes')
    if len(car_text) < counter:
        break
    else:
        page += 1
```

```python
columns = ['StockType', 'Title', 'Exterior', 'Interior', 'Transmission', 'Drivetrain', 'Price', 'Mileage']
```

```python
car_df = pd.DataFrame(cars, columns=columns)
```
